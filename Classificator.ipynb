```python
import extractor as ext
from sklearn.model_selection import train_test_split
import keras
from sklearn.metrics import confusion_matrix
keras.__version__
```

```python
X,y = ext.readFiles()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
from keras import models
from keras import layers
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D, Dropout
network = models.Sequential()
#input
network.add(Conv2D(64, kernel_size=3, padding="same", activation='relu', input_shape=(95, 25,3)))
network.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
network.add(Dropout(0.2))
#layer 1
network.add(Conv2D(64, kernel_size=3, padding="same",activation='relu'))
network.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
network.add(Dropout(0.2))
#layer 2
network.add(Conv2D(64, kernel_size=3, padding="same",activation='relu'))
network.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
network.add(Dropout(0.2))
#layer 3
network.add(Conv2D(64, kernel_size=3, padding="same",activation='relu'))
network.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
network.add(Dropout(0.2))
#layer 4
network.add(Conv2D(64, kernel_size=3, padding="same",activation='relu'))
network.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
network.add(Dropout(0.2))
#layer 5
network.add(Conv2D(64, kernel_size=3, padding="same",activation='relu'))
network.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
network.add(Dropout(0.2))
#layer 6
network.add(Conv2D(64, kernel_size=3, padding="same",activation='relu'))
network.add(MaxPooling2D(pool_size=(2, 2), padding="same"))
network.add(Dropout(0.2))

#output
network.add(Flatten())
network.add(layers.Dense(5, activation='softmax'))
```

```python
network.compile(optimizer='adam', 
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
train_images = X_train.astype('float32') / 255
test_images = X_test.astype('float32') / 255
```

```python
from keras.utils import to_categorical
train_labels = to_categorical(y_train)
test_labels = to_categorical(y_test)
```

```python
history = network.fit(train_images, train_labels, epochs=350, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
```

```python
print('test_acc:', round(test_acc,4)*100)
```

```python
print(str(test_labels[173]))
img = test_images[173]
img_tensor = image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)

plt.imshow(img_tensor[0])
plt.show()

print(img_tensor.shape)
```

```python
layer_outputs = [layer.output for layer in network.layers[:16]] # Extracts the outputs of the top 12 layers
activation_model = models.Model(inputs=network.input, outputs=layer_outputs) # Creates a model that will return these outputs, given the model input
activations = activation_model.predict(img_tensor) # Returns a list of five Numpy arrays: one array per layer activation

layer_names = []
for layer in network.layers[:16]:
    if "conv" in layer.name:
        layer_names.append(layer.name) # Names of the layers, so you can have them as part of your plot
    
images_per_row = 32
for layer_name, layer_activation in zip(layer_names, activations): # Displays the feature maps
    i=1
    n_features = layer_activation.shape[-1] # Number of features in the feature map
    sizeX = layer_activation.shape[1] #The feature map has shape (1, size, size, n_features).
    sizeY = layer_activation.shape[2]
    n_cols = n_features // images_per_row # Tiles the activation channels in this matrix
    display_grid = np.zeros((sizeX * n_cols, images_per_row * sizeY))
    for col in range(n_cols): # Tiles each filter into a big horizontal grid
        for row in range(images_per_row):
            channel_image = layer_activation[0,:, :,col * images_per_row + row]
            
            channel_image -= channel_image.mean() # Post-processes the feature to make it visually palatable
            y = channel_image.std() if channel_image.std() != 0 else 1 
            channel_image /= y
            channel_image *= 64
            channel_image += 128
            channel_image = np.clip(channel_image, 0, 255).astype('uint8')
            display_grid[col * sizeX : (col + 1) * sizeX, # Displays the grid
                         row * sizeY : (row + 1) * sizeY] = channel_image
    scale = 1. / sizeX * 10*i
    i = (i+1)*2
    plt.figure(figsize=(scale * display_grid.shape[1],
                        scale * display_grid.shape[0]))
    plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
```
